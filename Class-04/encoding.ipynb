```python
import numpy as np
from sklearn.preprocessing import OneHotEncoder
```

```python
documents = [
    "people watch movie",
    "people watch cricket",
    "people like movie",
    "people like cricket",
]
```

```python
tokens = [sentences.lower().split() for sentences in documents]
```

```python
tokens
```

```python
all_words = [[word] for sentences in tokens for word in sentences]
```

```python
all_words
```

```python
encoder = OneHotEncoder(sparse_output=False).fit(all_words)
```

```python
encoder.categories_
```

```python
print("Vocabulary:", encoder.categories_[0])
```

```python
tokens
```

```python
for sentences in tokens:
        encoded_sentence = encoder.transform([[word] for word in sentences])
        print("\nsentence:", sentences)
        print("encoded:\n", encoded_sentence) 
```
